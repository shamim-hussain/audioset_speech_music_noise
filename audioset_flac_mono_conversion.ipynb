```python
from pathlib import Path

source_dir = Path(r'G:\datasets\audioset')
dest_dir= Path(r'G:\datasets\audioset_flac')
```

```python
wav_files = list(source_dir.rglob('*.wav'))
len(wav_files)
```

```python
from tqdm import tqdm

for f_in in tqdm(wav_files):
    f_out = dest_dir/f_in.parent.relative_to(source_dir)/(f_in.stem+'.flac')
    f_out.parent.mkdir(exist_ok=True, parents=True)
    
    f_in = str(f_in)
    f_out = str(f_out)
    
    !ffmpeg -i "$f_in" -ac 1 -hide_banner -loglevel error "$f_out" 
```

```python
import pandas as pd

annotations_file =  dest_dir/'annotations.csv'
df_annotations = pd.read_csv(annotations_file).set_index('ytid')
df_annotations
```

```python
df_annotations['file_path'] = df_annotations.wav_file_path.str[:-4] + '.flac'
df_annotations
```

```python
df_annotations = df_annotations.drop(columns=['wav_file_path'])
df_annotations
```

```python
df_annotations.reset_index().to_csv(dest_dir/'annotations.csv', index=False)
```
