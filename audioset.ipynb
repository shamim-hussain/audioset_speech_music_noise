# Build Superclasses

## Load ontology

```python
import pandas as pd
import numpy as np

df_onto = pd.read_json('ontology/ontology.json')
df_onto.head()
```

Set id as index.

```python
df_onto.set_index('id',inplace=True)
df_onto
```

## View Main Classes

Speech

Music

```python
id_music = r'/m/04rlf'
df_onto.loc[id_music]
```

Human Sound

```python
id_human = r'/m/0dgw9r'
df_onto.loc[id_human]
```

Human Sounds

```python
print(f'Human sounds: Total {len(df_onto.loc[id_human])} child classes\n')
for i, id in enumerate(df_onto.loc[id_human].child_ids):
    print(f'[{i+1}]')
    print(df_onto.loc[id])
    print()
```

```python
print(f'Human voice sounds: Total {len(df_onto.loc["/m/09l8g"])} child classes\n')
for i, id in enumerate(df_onto.loc['/m/09l8g'].child_ids):
    print(f'[{i+1}]')
    print(df_onto.loc[id])
    print()
```

### Superclasses

```python
def get_all_subclasses(lb):
    lst=[]
    lst.append(lb)
    for c in df_onto.loc[lb,'child_ids']:
        lst.extend(get_all_subclasses(c))
    return lst

speech_classes = get_all_subclasses(id_speech)
music_classes = get_all_subclasses(id_music)
human_classes = get_all_subclasses(id_human)

noise_classes = list(cls for cls in df_onto.index if cls not in human_classes and cls not in music_classes)
```

```python
print('Speech Classes:\n')
for cls in speech_classes:
    print(df_onto.loc[cls,'name'])
```

```python
print('Music Classes:\n')
for cls in music_classes:
    print(df_onto.loc[cls,'name'])

print()
```

```python
print('Human sound Classes:\n')
for cls in human_classes:
    print(df_onto.loc[cls,'name'])
```

```python
print('Noise Classes:\n')
for cls in noise_classes:
    print(df_onto.loc[cls,'name'])
```

### Test exclusivity

```python
assert len(set(speech_classes).intersection(set(music_classes))) == 0
assert len(set(speech_classes).intersection(set(noise_classes))) == 0
assert len(set(noise_classes).intersection(set(music_classes))) == 0
```

### Save

```python
def get_superclass(cls):
    cls
    if cls in speech_classes:
        return 'speech'
    elif cls in music_classes:
        return 'music'
    elif cls in noise_classes:
        return 'noise'
    else:
        return 'nonspeech_human_sound'

df_onto['superclass']=list(map(get_superclass, df_onto.index))
df_onto
```

```python
import os
os.makedirs('processed',exist_ok=True)

df_onto.to_pickle('processed/ontology_with_superclass.pkl')
# df_onto.reset_index().to_json('processed/ontology_with_superclass.json',orient='records', indent=4)
```

Optionally save superclasses

```python
import yaml
super_classes = dict(
    speech = list({'id':id, 'name':df_onto.loc[id,'name']} for id in speech_classes),
    music = list({'id':id, 'name':df_onto.loc[id,'name']} for id in music_classes),
    noise = list({'id':id, 'name':df_onto.loc[id,'name']} for id in noise_classes),
    human = list({'id':id, 'name':df_onto.loc[id,'name']} for id in human_classes),
)

with open('processed/superclasses.yaml', 'w') as file:
    yaml.dump(super_classes, file)
```

# Apply superclass and combine datasets

## Load data

Optionally load ontology with superclass

```python
import pandas as pd
import numpy as np

df_onto = pd.read_pickle('processed/ontology_with_superclass.pkl')
```

Check csv file contents

```python
with open('eval_segments.csv','r') as file:
    for i, line in zip(range(10), file):
        print(line)
```

```python
df_eval = pd.read_csv('eval_segments.csv', sep=', ', engine ='python', skiprows=2)
df_balanced = pd.read_csv('balanced_train_segments.csv', sep=', ', engine ='python', skiprows=2)
df_unbalanced = pd.read_csv('unbalanced_train_segments.csv', sep=', ', engine ='python', skiprows=2)
df_eval
```

## Combine

```python
df_eval['dataset_split'] = 'eval'
df_balanced['dataset_split'] = 'train_balanced'
df_unbalanced['dataset_split'] = 'train_unbalanced'

df_audioset = pd.concat([df_eval,df_balanced,df_unbalanced])
df_audioset
```

## Assign superclasses

```python
onto_super = df_onto.superclass
def get_plausible_superclass(labels):
    labels = labels[1:-1].split(',')
    supers = set(onto_super[lb] for lb in labels)
    if len(supers) == 1:
        super_cls ,= supers
    else:
        super_cls = 'other'
    return super_cls 

df_audioset['plausible_superclass'] = df_audioset.positive_labels.apply(get_plausible_superclass)
df_audioset
```

Rename columns and reindex

```python
df_audioset.columns = ['ytid'] + list(df_audioset.columns[1:])
df_audioset.set_index('ytid',inplace=True)
df_audioset
```

Count plausible superclasses

```python
print('eval:')
print(df_audioset[df_audioset.dataset_split=='eval'].plausible_superclass.value_counts())
print()

print('train_balanced:')
print(df_audioset[df_audioset.dataset_split=='train_balanced'].plausible_superclass.value_counts())
print()

print('train_unbalanced:')
print(df_audioset[df_audioset.dataset_split=='train_unbalanced'].plausible_superclass.value_counts())
print()

print('overall:')
print(df_audioset.plausible_superclass.value_counts())
```
